# Seaborn data visualization

```python
# from https://youtu.be/PQUbwZrgdSE

import pandas as pd
import seaborn as sns
import os
%matplotlib inline
os.chdir("D://Docs//PyProjects//ml//lessons//seaborn")
```

```python
iris = pd.read_csv("iris.csv")
```

```python
iris.head()
```

```python
iris.Species.value_counts()
```

```python
sns.jointplot(x="Sepal length", y="Sepal width", data=iris, size=5)
```

```python
sns.jointplot(x="Petal length", y="Petal width", data=iris)
```

# Facet grid Plots

```python
import matplotlib.pyplot as plt
sns.FacetGrid(iris, hue="Species", size=5)\
    .map(plt.scatter, "Sepal length", "Sepal width")\
    .add_legend()
```

# Boxplot

```python
sns.boxplot(x="Species", y="Petal length", data = iris)
```

```python
sns.boxplot(x="Species", y="Petal length", data = iris)
sns.stripplot(x="Species", y="Petal length", data = iris, jitter=True, edgecolor="gray")
```

# Violin Plot

```python
sns.violinplot(x="Species", y= "Petal length", data=iris)
```

# Pairplot

```python
sns.pairplot(data=iris, kind = "scatter")
```
